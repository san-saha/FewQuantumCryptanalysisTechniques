```python
def init(row0, row1, row2, row3, base):
  A0,A1,A2,A3 = row0.split()
  B0,B1,B2,B3 = row1.split()
  C0,C1,C2,C3 = row2.split()
  D0,D1,D2,D3 = row3.split()

  A0=int(A0, base)
  B0=int(B0, base)
  C0=int(C0, base)
  D0=int(D0, base)

  A1=int(A1, base)
  B1=int(B1, base)
  C1=int(C1, base)
  D1=int(D1, base)

  A2=int(A2, base)
  B2=int(B2, base)
  C2=int(C2, base)
  D2=int(D2, base)

  A3=int(A3, base)
  B3=int(B3, base)
  C3=int(C3, base)
  D3=int(D3, base)

  return A0,A1,A2,A3,B0,B1,B2,B3,C0,C1,C2,C3,D0,D1,D2,D3
```

```python
def one_arx(a, b, c, d, n, shift):
  highest_num = (1 << n) -1
  a += b
  # a &= 0xffffffff
  a &= highest_num
  d ^= a
  d = (((d) << (shift)) | ((d) >> (n - (shift))))
  # d &= 0xffffffff
  d &= highest_num
  return a,b,c,d
```

```python
def add_original(ori_list, modified_list):
  highest_num = (1 << n) -1

  new_list = []
  for i, j in zip(ori_list, modified_list):
    j += i
    j &= highest_num
    new_list.append(j)

  return new_list
```
